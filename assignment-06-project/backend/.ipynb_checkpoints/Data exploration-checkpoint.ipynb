```python
sample_data = %sql select * from WDIData limit 5;
display(sample_data)
```

```python
# select countries
country_names = %sql select DISTINCT Country_Name from WDIData order by Country_Name asc;
print(len(country_names))
display(country_names)
```

```python
# select indicators
indicator_names = %sql select DISTINCT Indicator_Name from WDISeries order by Indicator_Name asc;
print(len(indicator_names))
display(indicator_names)
```

```python
topics = %sql select TOPIC, count(Indicator_Name) from WDISERIES group by topic order by topic asc;
display(topics)
```

```python
env_topics = %sql select TOPIC, count(Indicator_Name) from WDISERIES where topic like 'Environment%' group by topic order by topic asc;
display(env_topics)
```

```python
env_indicators = %sql select TOPIC, Indicator_Name from WDISERIES where topic like 'Environment%' order by topic asc;
print(len(env_indicators))
display(env_indicators)
```

```python
conn = sqlite3.connect('data/data.db')
```

```python
df_countries = pd.read_sql('''
select Country_Code as code, Table_Name as name, Income_Group as income_group from WDICountry where Income_Group is not null order by Income_Group asc;''', con = conn)
df_countries
```

```python
df_data = pd.read_sql('''
select WDISeries.Topic, WDIData.*
from WDIData 
JOIN WDISeries ON WDIData.Indicator_Name = WDISeries.Indicator_Name
where WDIData.Country_Name in ('Afghanistan', 'Canada', 'Ecuador', 'Germany') 
-- and WDIData.Indicator_Name in ('Access to electricity (% of population)', 'GDP growth (annual %)', 'Net foreign assets (current LCU)', 'Secure Internet servers')
and WDISeries.Topic like 'Environment: Agricultural production'
''', con = conn)
df_data
```

```python
import matplotlib.pyplot as plt
ax = plt.gca()
df_data[(df_data['Indicator_Name'] == 'Access to electricity (% of population)') & (df_data['Country_Name'] == 'Germany')].plot(x ='Year', y='Value', kind = 'scatter', color = 'red', ax = ax)
df_data[(df_data['Indicator_Name'] == 'Access to electricity (% of population)') & (df_data['Country_Name'] == 'Afghanistan')].plot(x ='Year', y='Value',  kind = 'scatter', color = 'blue', ax = ax)

plt.show()
```

## Task 1 - (c) Solution

### Research question

#### How are the Price, Owners and Positive ratings correlated in the steam games dataset?

Build a visual representation of the connection between 3 factors namely: Price, Owners and Positive rating.
The following columns will be used from the dataset: `price`, `positive_rating`, `owners`.

The visualization will be a sankey diagram. With 3 types of nodes and each type of node will have a different color (total 3 colors). The nodes will be connected with edges and width of the each connection will be proportional to the value of edge. The value of edge will be the number of games for the `source` and `target`.

As the `price` data is numeric, it will be group in ranges as follows: [<5, 5-20, 20-100, 100-250, 250+].
As the `positive_ratings` is numeric, it will be group in ranges as follows: [<1k, 1k-10k, 10k-100k, 100k-500k, 500k-2M, 2M+].

## Task 1 - (d) Solution

```python
import numpy as np

# Convert `price` column into range of values.
bins = [-1, 5, 20, 100, 250, df['price'].max()]
price_names = ['<5', '5-20', '20-100', '100-250', '250+']
df['price-range'] = pd.cut(df['price'], bins, labels=price_names)

# Convert `rating` column into range of values.
rating_bins = [-1, 1000, 10000, 100000, 500000, 2000000, 2000000*100]
rating_names = ['<1k', '1k-10k', '10k-100k', '100k-500k', '500k-2M', '2M+']

df['rating-range'] = pd.cut(df['positive_ratings'], rating_bins, labels=rating_names)

# `df_processed` will contain binned data and 3 columns. Each row is on
df_processed = df[["price-range", "owners","rating-range"]]
owners_names = list(df_processed["owners"].unique())
display(df_processed)
```

```python
# Change owners column to readable text.
new_owners_names = [{'old':'0-20000','new':'0-20k'},
                    {'old':'20000-50000','new':'20k-50k'},
                    {'old':'50000-100000','new':'50k-100k'},
                   {'old':'100000-200000','new':'100k-200k'},
                   {'old':'200000-500000','new':'200k-500k'},
                   {'old':'500000-1000000','new':'500k-1M'},
                   {'old':'1000000-2000000','new':'1M-2M'},
                   {'old':'2000000-5000000','new':'2M-5M'},
                   {'old':'5000000-10000000','new':'5M-10M'},
                   {'old':'10000000-20000000','new':'10M-20M'},
                   {'old':'20000000-50000000','new':'20M-50M'},
                   {'old':'50000000-100000000','new':'50M-100M'},
                   {'old':'100000000-200000000','new':'100M-200M'}]

sorted([r["old"] for r in new_owners_names]) == sorted(owners_names) 

old_names = [r["old"] for r in new_owners_names]
new_names = [r["new"] for r in new_owners_names]

df_processed.is_copy = False
df_processed.owners.replace(old_names,new_names, inplace=True)
df_processed
```

```python
# Group data

# df1 will contain links between `price-range` and `owners`
df1 = pd.DataFrame({'value' : df_processed.groupby(["price-range","owners"]).size()}).reset_index()
df1.rename(columns = {'price-range':'source', 'owners':'target'}, inplace = True)
print("links between `price-range` and `owners`")
display(df1)
print("*"*100)

# df1 will contain links between `owners` and `ratings-range`
df2 = pd.DataFrame({'value' : df_processed.groupby(["owners","rating-range"]).size()}).reset_index()
df2.rename(columns = {'owners':'source', 'rating-range':'target'}, inplace = True)
print("links between `owners` and `rating-range`")
display(df2)

# Merge the dataframes
df3 = pd.concat([df1, df2], sort=False)
df3.reset_index(drop=True, inplace=True)
df3 = df3[df3["value"] != 0]
```

```python
# Create nodes
nodes = [{'type': 'price', 'name': node} for node in sorted(list(df_processed["price-range"].unique()),reverse=True)]
nodes = nodes + [{'type': 'owners', 'name': node} for node in sorted(list(df_processed["owners"].unique()),reverse=True)]
nodes = nodes + [{'type': 'rating', 'name': node} for node in sorted(list(df_processed["rating-range"].unique()),reverse=True)]

nodes
```

```python
# Convert links to json.
import json
edges = json.loads(df3.to_json(orient="records"))
edges[:]
```

```python
# Process nodes. Change source, target to index from the node list
for index, node in enumerate(nodes):
    df3.replace(node['name'], index, inplace=True)

edges = json.loads(df3.to_json(orient="records"))
edges[:]
graph = {
    'nodes' : nodes,
    'links': edges
}

# Writing to graph.json
with open("graph.json", "w") as outfile:
    json_object = json.dumps(graph, indent = 4)
    outfile.write(json_object)
```
